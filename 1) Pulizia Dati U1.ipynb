```python
def pulizia(data_sistemare, prima_data, seconda_data, ora_inizio, ora_fine, edificio = U1, peso1 = 1, peso2 = 1):
    conta = ora_inizio
    while conta <= ora_fine :
        primo_valore = float(edificio[(edificio.ORA == conta) & (edificio.DATA == prima_data)]['CONSUMO_ATTIVA_PRELEVATA'])
        secondo_valore = float(edificio[(edificio.ORA == conta) & (edificio.DATA == seconda_data)]['CONSUMO_ATTIVA_PRELEVATA'])
        media = round(((primo_valore*peso1) + (secondo_valore*peso2)) / (peso1+peso2),2)
        edificio['CONSUMO_ATTIVA_PRELEVATA'].loc[(edificio.ORA == conta) & (edificio.DATA == data_sistemare)] = media
        if '45' in str(conta):
            conta += 5500
        else:
            conta += 1500
```

# 5) AGGREGAZIONE:

```python
U1['consumo15']=U1['CONSUMO_ATTIVA_PRELEVATA']*15
```

```python
def boh(ora):
    if ora==0 or ora==1500 or ora==3000 or ora==4500 :
        return 0 
    else:
        return(ora//10000)
```

```python
U1['Tempo']=U1['ORA'].apply(boh)
```

```python
U1_agg_ore = pd.DataFrame(U1[['DATA', 'Tempo', 'consumo15']].groupby(['DATA', 'Tempo']).sum()['consumo15']).reset_index()
U1_agg_ore.rename(columns={'Tempo': 'ORA', 'consumo15' : 'CONSUMO_KWH_U1'}, inplace = True)

U1_agg_giorni = pd.DataFrame(U1_agg_ore.groupby(['DATA']).sum()['CONSUMO_KWH_U1']).reset_index()
```
